```python
import torch

# テンソル生成
x = torch.tensor([1, 2, 3])
y = torch.ones(3)
z = torch.zeros(3)
print("x:", x)
print("y:", y)
print("z:", z)

# 形状指定で初期化
a = torch.empty(2, 3)
b = torch.rand(2, 3)
print("empty tensor a:\n", a)
print("random tensor b:\n", b)

# NumPyと相互変換
import numpy as np
np_array = np.array([[1, 2], [3, 4]])
tensor_from_np = torch.from_numpy(np_array)
back_to_np = tensor_from_np.numpy()
print("Tensor from NumPy:\n", tensor_from_np)
print("Back to NumPy:\n", back_to_np)

# 演算例
t1 = torch.tensor([[1., 2.], [3., 4.]])
t2 = torch.tensor([[5., 6.], [7., 8.]])
print("加算:\n", t1 + t2)
print("行列積:\n", torch.matmul(t1, t2.T))
```
